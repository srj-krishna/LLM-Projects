```python
import lilac as ll

signal = ll.signals.ConceptSignal(
  namespace='lilac',
  concept_name='positive-sentiment',
  embedding='gte-small')

# Signals take an iterable of inputs, and return a list of items that match the shape of the input.
result = list(signal.compute(['This was the best movie I have ever seen in my entire life!']))

print(result)
```

```python
# Load a concept
concept = ll.DiskConceptDB.get('local', 'positive-product-reviews')

# Print the examples in the concept
print(concept.data)
```

```python
# Add a new example to the concept
train_data = [
  ll.ExampleIn(label=True, text='This product is the best!'),
]
ll.DiskConceptDB.edit(
  'local', 'positive-product-reviews',
  ll.ConceptUpdate(insert=train_data))
```

```python
# Remove an example from the concept
ll.DiskConceptDB.edit(
  'local', 'positive-product-reviews',
  ll.ConceptUpdate(remove=['d86e4cb53c70443b8d8782a6847f4752']))
```

```python
# Delete a concept
ll.DiskConceptDB.remove('local', 'positive-product-reviews')
```
